```python
# Select only numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate Q1 and Q3 for each numerical column
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)

# Calculate IQR for each numerical column
IQR = Q3 - Q1
print(IQR)
# Determine bounds for each column
lower_bounds = Q1 - 1.5 * IQR
upper_bounds = Q3 + 1.5 * IQR

# Create a boolean mask to filter out rows with outliers in any numerical column
mask = ~(
    (df[numerical_cols] < lower_bounds) | (df[numerical_cols] > upper_bounds)
).any(axis=1)

# Apply the mask to get cleaned DataFrame
df = df[mask]
df.info()
```
